assume columns are independent and apply linear regression . last column predict using other columns . then split data into training and testing

```python
def splitTrainTest(df_stock , p=0.8):
  y = df_stock['Close_forcast']
  df_stock = df_stock.drop(columns = 'Close_forcast') # X
  print(df_stock.shape)
  n = df_stock.shape[0]
  ntr = int(n*p)
  x_Train , y_Train = df_stock[:ntr] , y[:ntr]
  x_Test , y_Test = df_stock[ntr:] , y[ntr:]
  print('Total - ' , n)
  print('Training : ' , x_Train.shape , y_Train.shape )
  print('Testing : ' , x_Test.shape , y_Test.shape )
  return x_Train , y_Train , x_Test , y_Test

xTr ,yTr , xTe , yTe = splitTrainTest(df_stock)
```

Sklearn

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , r2_score

model = LinearRegression()
model.fit(xTr , yTr)
print('LR Coefficients : /n',model.coef_)
print('LR Intercept : \n' , model.intercept_) #bias

```

```python
#testing
y_tr_pred = model.predict(xTr)
y_te_pred = model.predict(xTe)
print("Training R-Squared : " , round(r2_score(yTr , y_tr_pred) , 2))
print("Testing R-Squared : " , round(r2_score(yTe , y_te_pred) , 2))

#prediction
df_pred = pd.DataFrame(yTe.values , columns = ['Actual'])
df_pred['Predicted'] = y_te_pred
df_pred[['Actual', 'Predicted']].plot()
plt.show()
```
