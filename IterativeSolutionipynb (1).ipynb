Gradient Descent


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head(5)
```

```python
# X = hours , t = Scores and converting them to numpy array
# X design metrix
X_ori = np.array(df['Hours']).reshape(-1,1) # size N*1
t = np.array(df['Scores']).reshape(-1,1) # size N*1
print(X_ori.shape , t.shape)
```

```python
plt.scatter(X_ori,t)
plt.title("Data")
plt.ylabel("x (input ): Hours")
plt.xlabel("t (targets ): Scores")
plt.show()
```

```python
N = X_ori.shape[0]
D = X_ori.shape[1]
print('No of instances: ',N, '\nNo of Features : ' ,D)
```

```python
#Append 1 to remove a separate bias term
X = np.append(X_ori , np.ones((N,1)) , axis = 1) # both same
# X = np.append(X_ori , np.ones_like(X_ori) , axis = 1 ) # both same
D = D + 1
```

```python

w = np.zeros((D,1))
print(X.shape , t.shape , w.shape)
```

```python
#lr = 0.01 #learning rate
lr = 0.01 #learning rate change and try again
loss = []
nepoches = 30
for i in range(nepoches):
  y = X.dot(w)
  dldw = (X.T).dot(y-t)/N
  w = w - lr * dldw
  loss.append(((y-t)**2).mean())
```

```python
#plot loss
fig = plt.figure(figsize = (8,6))
plt.plot([i for i in range(nepoches)],loss,'r-')
plt.xlabel('Number of iterations')
plt.ylabel('loss / Cost')
plt.show()
```

```python
#predictions for training data
y = np.dot(X,w)
#plot Predictions
fig = plt.figure(figsize = (8,6))
plt.scatter(X_ori,t)
plt.plot(X_ori,y,'r-') # X and predictions
plt.title("Hours vs Precentage")
plt.ylabel("x (input ): Hours")
plt.xlabel("t (targets ): Scores")
plt.show()
```

```python
#Comapring TRUE values to our predictions
print(t.shape , y.shape)
Compare_df = pd.DataFrame({'Actual':t.reshape(-1),'Predicted':y.reshape(-1)})
print(Compare_df)
```
